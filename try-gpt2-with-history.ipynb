### Text generation

```python
from transformers import pipeline, set_seed
import warnings
warnings.filterwarnings("ignore")
```

```python
class MemoryChatbot:
    def __init__(self):
        self.generator = pipeline('text-generation', model='gpt2')
        set_seed(42)
        self.memory = []

    def respond(self, user_input):
        # Store user input in memory
        self.memory.append(user_input)
        
        # Generate response based on the conversation history
        context = " ".join(self.memory[-3:])  # Consider last 3 turns for context
        response = self.generator(context, max_length=60, num_return_sequences=1)[0]['generated_text']
        
        # Store response in memory
        self.memory.append(response)
        
        return response
```

```python
# Example usage
chatbot = MemoryChatbot()
print(chatbot.respond("Why women tend to live longer than men?"))
```

```python
print(chatbot.respond("What are some factors contributing to this?"))
```

```python
print(chatbot.respond("Do lifestyle choices play a role as well?"))
```
